Importação das bibliotecas do projeto

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("ggplot")
```

Leitura do arquivo 'Iris.csv' - apresentado em formato de tabela pela Pandas

```python
planta = pd.read_csv("Iris.csv")
```

Exibição das linhas iniciais da base de dados com a função *head()* - se nenhum valor for passado, ela exibe as 5 primeiras linhas por padrão

```python
planta.head()
```

Utilizando a função *describe()* para retornar o resumo das variáveis numéricas da base de dados com informações relevantes sobre os dados, como média ou desvio padrão, que poderão ser utilizadas para gerar estatísticas

```python
planta.describe()
```

A função *dtypes* retorna os tipos de dados

```python
planta.dtypes
```

Utilizando o comando *shape* da Pandas para retornar uma tupla com os valores de tamanho da base de dados, sendo respectivamente: a quantidade de linhas e colunas da tabela

```python
planta.shape
```

Utilizando o atributo *columns* da Pandas para renomear as colunas da tabela, com intenção de facilitar o entendimento da base de dados, seguindo os passos:
1. Informar os novos nomes em uma lista
2. Armazená-los como se fosse uma variável

```python
planta.columns = ["id", "sepala_comprimento", "sepala_largura", "petala_comprimento", "petala_largura", "especie"]
```

Utilizando a *head* novamente para verificar a alteração no nome das colunas

```python
planta.head()
```

A função *is_null()* retorna as linhas em que há valores nulos na coluna especificada

```python
planta[planta["sepala_comprimento"].isnull()].head()
```

A função *value_counts* retorna a contagem de valores de uma coluna específica

```python
planta["especie"].value_counts()
```

Utilizando a Pandas para criar o gráfico



*   kind = "tipo do gráfico"
*   figsize = (tamanho do gráfico)
*   grid = linhas de grade (True/False)
*   rot = grau de rotação dos dados que o eixo X deve ter
*   color = "cor do gráfico"

```python
planta["especie"].value_counts().head(10).plot(kind = "bar", figsize = (11,5), grid = False, rot = 0, color = "orange")
```

Utilizando MatplotLib para deixar o gráfico mais agradável

```python
planta["especie"].value_counts().head(10).plot(kind = "bar", figsize = (11,5), grid = False, rot = 0, color = "orange")

plt.title("Conjunto de dados de flores da Íris")
plt.xlabel("Classificação") # EIXO X - TIPOS DA ÍRIS
plt.ylabel("Quantidade") # EIXO Y - TOTAL DE TIPOS
plt.show() # EXIBIÇÃO DO GRÁFICO
```
